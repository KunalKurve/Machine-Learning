```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from catboost import CatBoostRegressor,CatBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import log_loss
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold, KFold
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import IsolationForest
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.compose import make_column_transformer, make_column_selector
from sklearn.pipeline import Pipeline
```

```python
train =pd.read_csv(r"D:/Machine Learning/Kaggle_COM_ML/Big_Mart_sales_Analytic_Vidhya/train.csv",index_col=0)
train
```

```python
test =pd.read_csv(r"D:/Machine Learning/Kaggle_COM_ML/Big_Mart_sales_Analytic_Vidhya/test.csv",index_col=0)
test
```

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

```python
train.drop(['Outlet_Identifier'],inplace=True,axis=1)
test.drop(['Outlet_Identifier'],inplace=True,axis=1)
```

```python
X_train=train.drop('Item_Outlet_Sales',axis=1)
y_train=train['Item_Outlet_Sales']
```

```python
impu_num=SimpleImputer(strategy='median').set_output(transform='pandas')
impu_cat=SimpleImputer(strategy='most_frequent').set_output(transform='pandas')

ct=make_column_transformer((impu_num,make_column_selector(dtype_include=np.number)),
                        (impu_cat,make_column_selector(dtype_include=object))).set_output(
                        transform='pandas')
```

```python
X_train=ct.fit_transform(X_train)
test=ct.transform(test)
```

```python
X_train = pd.get_dummies(X_train,drop_first=True)
test = pd.get_dummies(test,drop_first=True)
```

```python
cb=CatBoostRegressor()
cb.fit(X_train,y_train)
```

```python
y_pred = cb.predict(test)
df_test =pd.read_csv(r"D:/Machine Learning/Kaggle_COM_ML/Big_Mart_sales_Analytic_Vidhya/test.csv",index_col=0)
```

```python
y_pred[y_pred<0]=0
```

```python
submit = pd.DataFrame({"Item_Identifier":df_test.index,"Outlet_Identiifer":df_test["Outlet_Identifier"],
           "Item_Outlet_Sales":y_pred})
```

```python
submit.to_csv("Big_mart_cat_anlytics_vidhya.csv",index=False)
```
