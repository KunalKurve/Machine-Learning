```python
import os
import pandas as pd
import numpy as np
from statsmodels.graphics.tsaplots import plot_acf
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import adfuller
```

Analyzing the stationarity of the following time series

```python
os.chdir(r"D:\March 2024\PML\Datasets")
```

For AusGas.csv

```python
ausGas = pd.read_csv('AusGas.csv')
ausGas.head()
```

Stationarity Test: Augumented Dickey Fuller test

```python
result = adfuller(ausGas['GasProd'], maxlag=10)
print('p.value =', result[1])
```

```python
if result[1] < 0.05:
    print('The series is stationary')
else:
    print('The series is not stationary')
```

```python
plot_acf(ausGas['GasProd'], lags=30, alpha= None)
plt.show()
```

```python
plot_acf(ausGas['GasProd'], lags=30)
plt.show()
```

For FMAC-HPI_24420

```python
fmac = pd.read_csv('FMAC-HPI_24420.csv')
fmac.head()
```

```python
result = adfuller(fmac['NSA Value'] , maxlag=10)
print('p.value =', result[1])
```

```python
if result[1] < 0.05:
    print('The series is stationary')
else:
    print('The series is not stationary')
```

```python
diff_1 = fmac['NSA Value'].diff()
diff_1.dropna(inplace = True)
result = adfuller(diff_1, maxlag = 10)
print('p.value =', result[1])
```

```python
if result[1] < 0.05:
    print('The series is stationary')
else:
    print('The series is not stationary')
```

```python
diff_2 = diff_1.diff(1)
diff_2.dropna(inplace = True)
result = adfuller(diff_2, maxlag = 10)
print('p.value =', result[1])
```

```python
if result[1] < 0.05:
    print('The series is stationary')
else:
    print('The series is not stationary')
```

```python
plot_acf(fmac['NSA Value'], lags=30, alpha= None)
plt.show()
```

```python
plot_acf(fmac['NSA Value'], lags=30)
plt.show()
```
